```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

#     
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

#    
kernel = C(1.0, (1e-4, 1e1)) * RBF(0.5, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=1e-2, normalize_y=True)

#  
gp.fit(xdata[:, np.newaxis], ydata)

#     
xfit = np.linspace(0, 10, 1000)[:, np.newaxis]

#     
yfit, sigma = gp.predict(xfit, return_std=True)
dyfit = 2 * sigma  # 2*sigma ~ 95% 
```

```python
#  
plt.figure(figsize=(10, 5))
plt.errorbar(xdata, ydata, yerr=0.1, fmt='ok', lw=1, label='data')
plt.plot(xfit, yfit, '-b', label='prediction')
plt.fill_between(xfit.ravel(), yfit - dyfit, yfit + dyfit, color='blue', alpha=0.2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Gaussian Process Regression')
plt.legend()
plt.show()
```

```python
##     
from datetime import timedelta
def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return timedelta(hours=h, minutes=m, seconds=s)
```

```python
dt = pd.read_csv('https://raw.githubusercontent.com/jakevdp/marathon-data/master/marathon-data.csv',
                 converters={'split': convert_time, 'final': convert_time})
```

```python
dt.head()
```

```python
dt.info()
```

```python
dt.dtypes
```

```python
# Seaborn       
dt['split_sec'] = dt['split'].dt.total_seconds()
dt['final_sec'] = dt['final'].dt.total_seconds()
dt.head()
```

```python
dt.info()
```

```python
#       jointplot  
with sns.axes_style('white'):
    g = sns.jointplot(x="split_sec", y="final_sec", data=dt, kind='hex')
    #g = sns.jointplot(x="split_sec", y="final_sec", data=dt, kind='hist')
    g.ax_joint.plot(np.linspace(4000, 16000),np.linspace(8000,32000),':k')
```

```python
#             
#    
dt['split_frac'] = 1 - 2 * dt['split_sec'] / dt['final_sec']
dt.head()
```
