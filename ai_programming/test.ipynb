```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
data = pd.read_csv'data/FremontBridge.csv', index_col='Date', parse_dates=True date_format='%m%d%%Y')
data.head()
```

```python
data.info()
```

```python
data.index
```

```python
data.columns = ['Total','West', 'East']
```

```python
data.describe()
```

```python
data.dropna().describe()
```

```python
data.plot()
plt.ylabel('Hourly Bicycle Count');
```

```python
weekly = data.resample('W').sum()
weekly.plot(style=[':','--','-'])
plt.ylabel('Weekly bicycle count');
```

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count');
```

```python
daily.rolling(50, center=True,
              win_type='gaussian').sum(std=10).plot(style=[':', '--', '-']);
```

```python
import matplotlib
import os
import shutil

# matplotlib의 캐시 디렉토리 경로
cache_dir = matplotlib.get_cachedir()
font_cache_path = os.path.join(cache_dir, 'fontList.json')

# 폰트 캐시 파일 삭제
if os.path.exists(font_cache_path):
    os.remove(font_cache_path)
    print("폰트 캐시 파일이 삭제되었습니다.")
else:
    print("폰트 캐시 파일이 존재하지 않습니다.")
```
